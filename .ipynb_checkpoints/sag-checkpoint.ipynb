```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
```

### Crie um dashboard com gráficos que mostrem a faixa etária, o gênero, as profissões, o estado civil e a localidade dos clientes.

```python
sag = pd.read_excel('sag.xlsx')
sag.columns = sag.iloc[0]
clientes = sag.drop(sag.index[0])
clientes.columns = [c.replace(' ', '_') for c in clientes.columns]
```

```python
faixa_etaria = clientes.FAIXA
fe = faixa_etaria.value_counts().to_frame().reset_index().rename(columns = {"index" : "faixa", "FAIXA" : "qtd_faixa"})
sns.set(style="whitegrid")
ax = sns.barplot(x="qtd_faixa", y="faixa", data=fe, ci=None)
figure = ax.get_figure()    
ax.set(xlabel='Quantidade de clientes por faixa etária', ylabel='Faixa etária por clientes')
figure.suptitle('Faixa etária dos clientes', y=0.97, fontsize=18)
```

```python
genero = clientes.GÊNERO
gen = genero.value_counts().to_frame().reset_index().rename(columns = {"index" : "genero", "GÊNERO" : "qtd_genero"})
sns.set(style="whitegrid")
ax = sns.barplot(x="qtd_genero", y="genero", data=gen, ci=None)
figure = ax.get_figure()    
ax.set(xlabel='Quantidade  de clientes por gênero', ylabel='Gêneros')
figure.suptitle('Gênero dos clientes', y=0.97, fontsize=18)
```

```python
profissao = clientes.PROFISSÃO
prof = profissao.value_counts().to_frame().reset_index().rename(columns = {"index" : "profissao", "PROFISSÃO" : "qtd_profissao"})
sns.set(style="whitegrid")
ax = sns.barplot(x="qtd_profissao", y="profissao", data=prof, ci=None)
figure = ax.get_figure()    
ax.set(xlabel='Quantidade de clientes por profissão', ylabel='Profissões')
figure.suptitle('Profissão dos clientes', y=0.97, fontsize=18)
```

```python
estado_civil = clientes.ESTADO_CIVIL
ec = estado_civil.value_counts().to_frame().reset_index().rename(columns = {"index" : "ec", "ESTADO_CIVIL" : "qtd_ec"})
sns.set(style="whitegrid")
ax = sns.barplot(x="qtd_ec", y="ec", data=ec, ci=None)
figure = ax.get_figure()    
ax.set(xlabel='Quantidade de clientes por estado civil', ylabel='Estados civis')
figure.suptitle('Estado civil dos clientes', y=0.97, fontsize=18)
```

```python
cidade = clientes.CIDADE
ci = cidade.value_counts().to_frame().reset_index().rename(columns = {"index" : "ci", "CIDADE" : "qtd_ci"})
sns.set(style="whitegrid")
ax = sns.barplot(x="qtd_ci", y="ci", data=ci, ci=None)
figure = ax.get_figure()    
ax.set(xlabel='Quantidade de clientes por cidade', ylabel='Cidades')
figure.suptitle('Cidades dos clientes', y=0.97, fontsize=18)
```

### Qual o produto que representa o maior volume de vendas?

```python
vendas = pd.read_csv('vendas.csv')
vendas.rename(columns={'PREÇO UNIT':'PRECO_UNIT'}, inplace=True)
vendas['ano'] = pd.DatetimeIndex(vendas['DATA']).year
vendas['mes'] = pd.DatetimeIndex(vendas['DATA']).month
```

```python
volume = vendas.PRODUTO
v= volume.value_counts().to_frame().reset_index().rename(columns = {"index" : "produto", "PRODUTO" : "qtd_produto"})
sns.set(style="whitegrid")
ax = sns.barplot(x="qtd_produto", y="produto", data=v, ci=None)
figure = ax.get_figure()    
ax.set(xlabel='Quantidade de produtos vendidos', ylabel='Produtos')
figure.suptitle('Volume de vendas', y=0.97, fontsize=18)
```

### Qual o comportamento de vendas dos produtos ao longo do tempo?

```python
qm = vendas.groupby(['mes'])['QT'].agg('sum').to_frame().reset_index().rename(columns = {"QT" : "qtd_mes"})
ax = sns.lineplot(x="mes", y="qtd_mes", data=qm)
ax.set(xlabel='Meses', ylabel='Quantidade de produtos vendidos')
ax.set_title('Comportamento de vendas dos produtos em 2016', y=1.04, fontsize=18)
```
