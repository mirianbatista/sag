```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
sag = pd.read_excel('sag.xlsx')
sag.columns = sag.iloc[0]
clientes = sag.drop(sag.index[0])
clientes.columns = [c.replace(' ', '_') for c in clientes.columns]
clientes
```

```python
faixa_etaria = clientes.FAIXA
fe = faixa_etaria.value_counts().to_frame().reset_index().rename(columns = {"index" : "faixa", "FAIXA" : "qtd_faixa"})
sns.set(style="whitegrid")
ax = sns.barplot(x="qtd_faixa", y="faixa", data=fe, ci=None)
figure = ax.get_figure()    
ax.set(xlabel='Quantidade de clientes por faixa etária', ylabel='Faixa etária por clientes')
figure.suptitle('Faixa etária dos clientes', y=1.04, fontsize=18)
```

```python
genero = clientes.GÊNERO
gen = genero.value_counts().to_frame().reset_index().rename(columns = {"index" : "genero", "GÊNERO" : "qtd_genero"})
sns.set(style="whitegrid")
ax = sns.barplot(x="qtd_genero", y="genero", data=gen, ci=None)
figure = ax.get_figure()    
ax.set(xlabel='Quantidade  de clientes por gênero', ylabel='Gêneros')
figure.suptitle('Gênero dos clientes', y=1.04, fontsize=18)
```

```python
profissao = clientes.PROFISSÃO
prof = profissao.value_counts().to_frame().reset_index().rename(columns = {"index" : "profissao", "PROFISSÃO" : "qtd_profissao"})
sns.set(style="whitegrid")
ax = sns.barplot(x="qtd_profissao", y="profissao", data=prof, ci=None)
figure = ax.get_figure()    
ax.set(xlabel='Quantidade de clientes por profissão', ylabel='Profissões')
figure.suptitle('Profissão dos clientes', y=1.04, fontsize=18)
```
